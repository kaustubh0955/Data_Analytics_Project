**Preparing the Dataset**

```python
import pandas as pd
eight_st=pd.read_csv("/content/8_states.csv")
eight_st.head()
```

```python
eight_st.drop(0,axis=0,inplace=True)
```

```python
eight_st.head()
```

```python
eight_st.rename(columns={'CITY': 'City'},inplace=True)
eight_st.head()
```

```python
portland=pd.read_csv("/content/portland.csv")
portland.head()
```

```python
drop_colunm8=["PIc Included"	,"LANGUAGE"	,"SOURCE","H.Count"]
eight_st.drop(drop_colunm8,axis=1,inplace=True)
eight_st.head()
```

```python
drop_column=["Picture included"	,"Language","Reviewer Data","Helpful count"]
portland.drop(drop_column,axis=1,inplace=True)
portland.head()
```

```python
atlanta=pd.read_csv("/content/atlanta.csv")
atlanta.drop(drop_column,axis=1,inplace=True)
atlanta.head()
```

```python
indianpolis=pd.read_csv("/content/indianpolis.csv")
indianpolis.drop(drop_column,axis=1,inplace=True)
indianpolis.head()
```

```python
tacoma=pd.read_csv("/content/tacoma.csv")
tacoma.drop(drop_column,axis=1,inplace=True)
tacoma.head()
```

```python
data=pd.concat([eight_st,portland,atlanta,tacoma,indianpolis])
data['Rating'] = pd.to_numeric(data['Rating'], errors='coerce')  # 'coerce' will convert non-numeric values to NaN

data.reset_index(drop=True,inplace=True)
data.head()
```

```python
data["Rating"].sum()
```

```python
data.shape
```

```python
data.drop(["Unnamed: 0","Unnamed: 1"],axis=1,inplace=True)
```

```python
data.shape
```

```python
data.head()
```

```python
# @title Date vs count()

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  from matplotlib import pyplot as plt
  import seaborn as sns
  palette = list(sns.palettes.mpl_palette('Dark2'))
  counted = (series['Date']
                .value_counts()
              .reset_index(name='counts')
              .rename({'index': 'Date'}, axis=1)
              .sort_values('Date', ascending=True))
  xs = counted['Date']
  ys = counted['counts']
  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = data.sort_values('Date', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Date')
_ = plt.ylabel('count()')
```

*Location wise Average rating*

```python
df=pd.DataFrame(data)
city_avg_ratings = df.groupby('City')['Rating'].mean().reset_index()
city_avg_ratings['Rating'] = city_avg_ratings['Rating'].round(3)
# city_avg_ratings['Rating'] = city_avg_ratings['Rating'].apply(lambda x: x if x < 1e10 else float('nan'))

print(city_avg_ratings)
```

```python
city_avg_ratings.to_csv('city_avg_ratings.csv', index=False)
```

*Year wise average rating*

```python
data.dtypes
```

```python
data["Date"]
```

```python
df = pd.DataFrame(data)

# Convert 'Date' column to datetime with errors='coerce'
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Drop rows with NaT values in 'Date' column
df = df.dropna(subset=['Date'])

# Extract year from 'Date' column
df['Year'] = df['Date'].dt.year

# Group by 'City' and 'Year', calculate average rating
city_year_avg_ratings = df.groupby(['City', 'Year'])['Rating'].mean().reset_index()
```

```python
city_year_avg_ratings
```

```python
city_year_avg_ratings.to_csv('city_year_avg_ratings.csv', index=False)
```

***Location vs year number of views***

```python
location_year_number_ratings = df.groupby(['City', 'Year'])['Rating'].sum().reset_index()
```
